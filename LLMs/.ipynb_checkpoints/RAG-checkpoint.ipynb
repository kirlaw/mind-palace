**ÁÆÄÂçïÊªëÂä®Á™óÂè£‰ª£Á†ÅÁ§∫‰æãÔºö**

```python
MAX_TOKENS = 4000
RESERVE_TOKENS = 500   # ÁïôÁªôÂõûÂ§ç

def truncate_history(history, tokenizer):
    tokens = []
    new_history = []
    for turn in reversed(history):  # ‰ªéÊúÄËøëÂºÄÂßãÂæÄÂâç
        t = tokenizer(turn)
        if len(tokens) + len(t) > MAX_TOKENS - RESERVE_TOKENS:
            break
        tokens = t + tokens
        new_history.insert(0, turn)
    return new_history
```
